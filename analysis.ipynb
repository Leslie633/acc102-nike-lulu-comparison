## 1. Problem Definition & Target User

This notebook analyzes the financial performance of Nike and Lululemon over five fiscal years (2021–2025). The main questions are:
- Which company is more profitable (gross margin, net margin, ROA, ROE)?
- How have these metrics evolved over time?
- How does each company compare against industry benchmarks?

**Target users:** Accounting students, retail investors, and anyone interested in comparing two leading sportswear companies.

## 2. Dataset

- **Source:** Nike and Lululemon annual reports (Form 10-K) filed with the U.S. Securities and Exchange Commission (SEC), FY2021–FY2025.
- **Accessed:** April 2026
- **Key fields:** company, year, revenue, cost, expenses, assets, liabilities
- **Unit:** Billion USD
- **Reason for selection:** Publicly available, comparable business models, rich for ratio analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("nike_lulu.csv")
df.head()
```

## 3. Data Preparation & Metric Calculations

We compute:
- **Gross Profit** = Revenue – Cost
- **Gross Margin** = Gross Profit / Revenue × 100
- **Net Profit** = Gross Profit – Expenses
- **Net Margin** = Net Profit / Revenue × 100
- **Equity** = Assets – Liabilities
- **ROA** (Return on Assets) = Net Profit / Assets × 100
- **ROE** (Return on Equity) = Net Profit / Equity × 100

```python
# Perform calculations
df["gross_profit"] = df["revenue"] - df["cost"]
df["gross_margin"] = (df["gross_profit"] / df["revenue"]) * 100
df["net_profit"] = df["gross_profit"] - df["expenses"]
df["net_margin"] = (df["net_profit"] / df["revenue"]) * 100
df["equity"] = df["assets"] - df["liabilities"]
df["roa"] = (df["net_profit"] / df["assets"]) * 100
df["roe"] = (df["net_profit"] / df["equity"]) * 100

# Round to two decimals
df = df.round(2)
df.head()
```

## 4. Key Metrics for 2025

Filtering for the most recent year (2025) shows a side‑by‑side comparison.

```python
df_2025 = df[df["year"] == 2025]
key_cols = ["company", "revenue", "cost", "gross_profit", "gross_margin",
            "net_profit", "net_margin", "roa", "roe"]
df_2025[key_cols]
```

## 5. Industry Benchmark Comparison

Industry average gross margin = 45%, net margin = 12% (estimates). We classify each company as *Above* or *Below*.

```python
industry_gm = 45
industry_nm = 12

df_2025["gm_vs_industry"] = df_2025["gross_margin"].apply(lambda x: "Above" if x > industry_gm else "Below")
df_2025["nm_vs_industry"] = df_2025["net_margin"].apply(lambda x: "Above" if x > industry_nm else "Below")

df_2025[["company", "gross_margin", "gm_vs_industry", "net_margin", "nm_vs_industry"]]
```

## 6. Visual Analysis

### 6.1 Gross Margin Trend (2021–2025)

```python
plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots(figsize=(10, 5))

for company in ["Nike", "Lululemon"]:
    data = df[df["company"] == company]
    ax.plot(data["year"], data["gross_margin"], marker='o', label=company, linewidth=2)

ax.set_title("Gross Margin Trend (2021-2025)")
ax.set_xlabel("Year")
ax.set_ylabel("Gross Margin (%)")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

### 6.2 Revenue Trend (2021–2025)

```python
fig, ax = plt.subplots(figsize=(10, 5))
for company in ["Nike", "Lululemon"]:
    data = df[df["company"] == company]
    ax.plot(data["year"], data["revenue"], marker='s', label=company, linewidth=2)
ax.set_title("Revenue Trend (2021-2025)")
ax.set_xlabel("Year")
ax.set_ylabel("Revenue (Billion USD)")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

### 6.3 2025 ROA and ROE Comparison

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
x = df_2025["company"]
ax1.bar(x, df_2025["roa"], color="#1DA1F2", edgecolor="black")
ax1.set_title("ROA (2025)")
ax1.set_ylabel("ROA (%)")
ax2.bar(x, df_2025["roe"], color="#FF6B6B", edgecolor="black")
ax2.set_title("ROE (2025)")
ax2.set_ylabel("ROE (%)")
plt.show()
```

## 7. Key Findings

- **Lululemon** consistently has higher gross margin (~60%) than Nike (~44%).
- Lululemon’s net margin (~27%) is more than double Nike’s (~11%).
- ROE of Lululemon (~64%) is significantly higher than Nike’s (~25%), indicating better return for shareholders.
- Nike’s revenue ($53.5B in 2025) is over 4× larger than Lululemon’s ($12.0B), showing superior scale.
- Both companies exceed the industry average gross margin (45%).
